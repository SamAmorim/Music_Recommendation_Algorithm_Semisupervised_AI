```python
df_original['decade'] = (df_original['year'] // 10) * 10

# Calcular a média por década
mean_values = df_original.groupby('decade')[['acousticness', 'danceability', 'energy', 'instrumentalness']].mean().reset_index()

# Criar o gráfico
plt.figure(figsize=(12, 6))
for column in ['acousticness', 'danceability', 'energy', 'instrumentalness']:
    plt.plot(mean_values['decade'], mean_values[column], marker='o', label=column)

plt.title('Média de características musicais por década')
plt.xlabel('Década')
plt.ylabel('Média')
plt.xticks(mean_values['decade']) 
plt.legend()
plt.grid()
plt.show()
```
