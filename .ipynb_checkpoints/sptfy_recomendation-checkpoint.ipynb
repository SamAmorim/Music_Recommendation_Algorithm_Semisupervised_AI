```python
plt.hist2d(df_original['danceability'], df_original['energy'], bins=30, cmap='Blues')

# Adicionar título e rótulos
plt.title('Danceability vs Energy' )
plt.xlabel('energy')
plt.ylabel('danceability')
```

```python
df_original['decade'] = (df_original['year'] // 10) * 10

# Calcular a média por década
mean_values = df_original.groupby('decade')[['acousticness', 'danceability', 'energy', 'instrumentalness']].mean().reset_index()

# Criar o gráfico
plt.figure(figsize=(12, 6))
for column in ['acousticness', 'danceability', 'energy', 'instrumentalness']:
    plt.plot(mean_values['decade'], mean_values[column], marker='o', label=column)

plt.title('Média de características musicais por década')
plt.xlabel('Década')
plt.ylabel('Média')
plt.xticks(mean_values['decade']) 
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar o DataFrame para remover entradas sem décadas válidas
df_original = df_original[df_original['decade'].astype(bool)]

# Agrupar por década e calcular a média das características
mean_df = df_original.groupby('decade')[['acousticness', 'danceability', 'energy', 'instrumentalness', 'popularity']].mean().reset_index()

# Criar uma figura com um tamanho específico
plt.figure(figsize=(15, 10))

# Definir as características para análise
features = ['acousticness', 'danceability', 'energy', 'instrumentalness']

# Loop para criar gráficos de dispersão para cada característica em relação à popularidade média por década
for i, feature in enumerate(features):
    # Criar um subplot para cada característica
    plt.subplot(2, 2, i + 1)
    
    # Criar gráfico de dispersão
    sns.scatterplot(data=mean_df, x=feature, y='popularity', hue='decade', palette='viridis', alpha=0.7, s=100)
    
    # Adicionar uma linha de tendência
    sns.regplot(data=mean_df, x=feature, y='popularity', scatter=False, color='red')
    
    # Adicionar título e rótulos aos eixos
    plt.title(f'Relação entre {feature} e Popularidade (Média por Década)')
    plt.xlabel(feature)
    plt.ylabel('Popularidade')
    
    # Adicionar uma grade para melhor visualização
    plt.grid()

# Ajustar o layout para evitar sobreposição
plt.tight_layout()

# Exibir os gráficos
plt.show()
```
